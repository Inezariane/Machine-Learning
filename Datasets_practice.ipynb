```python
import pandas as pd
import numpy as np
import seaborn as sns 
from matplotlib import style
import plotly.express as px
import matplotlib.pyplot as plt


df=pd.read_csv('vehicles.csv')
df.head()
```

```python
def number_of_bins(data):
    n = len(data)
    k = 1 + 3.322 * np.log10(n)
    return int(np.ceil(k))

numb_bins = number_of_bins(df['proposed_purchase_price'])

plt.hist(df['proposed_purchase_price'], bins=numb_bins, color='blue', edgecolor='black')
plt.title('Histogram of Proposed Purchase Price')
plt.xlabel('Proposed Purchase Price')
plt.ylabel('Frequency')
plt.show()
```
